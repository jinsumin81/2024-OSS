### 선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
a=0
b=0

lr=0.03

epochs=2001
```

```python
n=len(x)

for i in range(epochs):
  y_pred = a*x+b
  error=y-y_pred

  a_diff = (2/n)*sum(-x * (error))
  b_diff = (2/n)*sum(-(error))

  a=a-lr * a_diff
  b=b-lr * b_diff

  if i % 100 == 0:
    print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i,a,b))
```

```python
y_pred=a*x+b

plt.scatter(x,y)
plt.plot(x, y_pred,'r')
plt.show()
```

###다중선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2,4,6,8])
x2 = np.array([1,2,1,6])
y = np.array([81, 93, 91, 97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2,y)
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.01

epochs = 2001
```

```python
n = len(x1)

for i in range(epochs):
    y_pred = a1 * x1 + a2 * x2 + b
    error = y - y_pred

    a1_diff = (2/n) * sum(-x1 * (error))
    a2_diff = (2/n) * sum(-x2 * (error))
    b_diff = (2/n) * sum(-(error))

    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff

    if i % 100 == 0:
        print("epoch=%d, 기울기1=%.4f, 기울기2=%.4f, 절편=%.4f" % (i, a1, a2, b))
```

```python
print("실제 점수:", y)
print("예측 점수:", y_pred)
```

###텐서플로우-선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()

model.add(Dense(1, input_dim=1, activation='linear'))

model.compile(optimizer='sgd', loss='mse')

model.fit(x,y, epochs=2000)
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x),'r')
plt.show()
```

```python
import tensorflow as tf

hour = 7

input_data = tf.constant([[hour]])
prediction = model.predict(input_data)[0][0]

print("%.f을 이동할 경우의 예상 금액은 %.02f원입니다" % (hour, prediction))
```

### 텐서플로우-다중선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([[2,0.5], [4, 0.8], [6,1], [8, 2.1]])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()

model.add(Dense(1, input_dim=2, activation='linear'))

model.compile(optimizer='sgd', loss='mse')

model.fit(x,y, epochs=2000)
```

```python
import tensorflow as tf

distance =7
minutes = 5

input_data = tf.constant([[distance, minutes]])
prediction = model.predict(input_data)[0][0]

print("%.f을 이동하고 %.f분을 움직이면 예상 금액은 %.02f원 입니다" % (distance, minutes, prediction))
```
